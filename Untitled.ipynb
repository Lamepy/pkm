## Functions

```python
def connect(url):
    page = requests.get(url)
    soup = bs.BeautifulSoup(page.text,"lxml")
    return soup
```

## Pickle file names

```python
pokemon_stats_df_picklename = 'pokemon_stats_df.pkl'
pokemon_moves_df_picklename = 'pokemon_moves_df.pkl'
pokemon_learnset_df_picklename = 'pokemon_learnset_df.pkl'
```

# Data Mining

## Pokemon Stats

```python
pokemon_stats_url = 'https://www.serebii.net/pokedex-gs/stat/all.shtml'

soup = connect(pokemon_stats_url)

stats_table_soup = soup.findAll('table',{'class':'dextable'})[0]

ranks = []
poke_index=[]
pokemon = []
main_type =[]
secondary_type =[]
hp=[]
attack=[]
defence = []
special_attack =[]
special_defence =[]
speed = []

for row in stats_table_soup.findAll('tr')[2:]:
    td = row.findAll('td')
    #Rank
    rank = td[0].get_text().split('/')[0].strip()
    if len(rank)>0:
        ranks.append(td[0].get_text().split('/')[0].strip())
    values =  row.findAll('td',{'class':'fooinfo'})
    if len(values)>0:
        #Pokedex Entry
        poke_index.append(values[0].get_text().strip())
        #Pokemon Name
        pokemon.append(values[2].get_text().strip().lower())
        
        #Main ans Secondary type
        types = values[3].findAll('a')

        if len(types)==1:
                main_type.append(types[0].get('href').replace('/pokedex-gs/','').replace('.shtml',''))
                secondary_type.append('None')
        if len(types)==2:
                main_type.append(types[0].get('href').replace('/pokedex-gs/','').replace('.shtml',''))
                secondary_type.append(types[1].get('href').replace('/pokedex-gs/','').replace('.shtml',''))
        
        #Hp
        hp.append(int(values[4].get_text()))
        
        #Attack
        attack.append(int(values[5].get_text()))
        
        #Defence
        defence.append(int(values[6].get_text()))
        
        #Special Attack 
        special_attack.append(int(values[7].get_text()))
        
        #Special Defence
        special_defence.append(int(values[8].get_text()))
        
        #Speed
        speed.append(int(values[9].get_text()))
        

pokemon_stats_df = pd.DataFrame()
pokemon_stats_df['rank']=ranks
pokemon_stats_df['poke_index']=poke_index
pokemon_stats_df['pokemon']=pokemon
pokemon_stats_df['main_type']=main_type
pokemon_stats_df['secondary_type']=secondary_type
pokemon_stats_df['hp']=hp
pokemon_stats_df['attack']=attack
pokemon_stats_df['defence']=defence
pokemon_stats_df['special_attack']=special_attack
pokemon_stats_df['special_defence']=special_defence
pokemon_stats_df['speed']=speed
pokemon_stats_df['total']=pokemon_stats_df['hp']+pokemon_stats_df['attack']+pokemon_stats_df['defence']+pokemon_stats_df['special_attack']+pokemon_stats_df['special_defence']+pokemon_stats_df['speed']


pokemon_stats_df.to_pickle(pokemon_stats_df_picklename)

pokemon_stats_df.head(5)
```

## Pokemon Moves

```python
moves_gen1_url = 'https://pokemondb.net/move/generation/1'
moves_gen2_url = 'https://pokemondb.net/move/generation/2'

soup_gen1 = connect(moves_gen1_url)
soup_gen2 = connect(moves_gen2_url)

gen1_moves_table_soup = soup_gen1.find('table',{'class':'data-table sticky-header block-wide'})
gen2_moves_table_soup = soup_gen2.find('table',{'class':'data-table sticky-header block-wide'})

move_name =[]
move_type =[]
move_category =[]
move_power=[]
move_accuracy =[]
move_pp =[]
move_effect =[]

for row in gen1_moves_table_soup.findAll('tr')[1:]:
    td=row.findAll('td')
    
    #Move Name
    move_name.append(td[0].get_text().lower())
    #Move_Type
    move_type.append(td[1].get_text().lower())
    #Move Category
    move_category.append(td[2].find('img').get('alt'))
    #Move Poer
    move_power.append(int(td[3].get_text().replace('—','0')))
    #Move Accuracy
    move_accuracy.append(td[4].get_text().replace('—','0'))
    #Move PP
    move_pp.append(td[5].get_text().replace('—','0'))
    #Move Effect
    move_effect.append(td[6].get_text())
    
for row in gen2_moves_table_soup.findAll('tr')[1:]:
    td=row.findAll('td')
    
    #Move Name
    move_name.append(td[0].get_text().lower())
    #Move_Type
    move_type.append(td[1].get_text().lower())
    #Move Category
    move_category.append(td[2].find('img').get('alt'))
    #Move Power
    move_power.append(int(td[3].get_text().replace('—','0')))
    #Move Accuracy
    move_accuracy.append(td[4].get_text().replace('—','0'))
    #Move PP
    move_pp.append(td[5].get_text().replace('—','0'))
    #Move Effect
    move_effect.append(td[6].get_text())
    
    
pokemon_moves_df =pd.DataFrame()
pokemon_moves_df['move_name']=move_name
pokemon_moves_df['move_type']=move_type
pokemon_moves_df['move_category']=move_category
pokemon_moves_df['move_power']=move_power
pokemon_moves_df['move_accuracy']=move_accuracy
pokemon_moves_df['move_pp']=move_pp
pokemon_moves_df['move_effect']=move_effect

pokemon_moves_df.to_pickle(pokemon_moves_df_picklename)

pokemon_moves_df.head(5)
```

## Pokemon Learnsets

```python
pokemon_learnset_url ='https://pokemondb.net/pokedex/game/gold-silver-crystal'

soup = connect(pokemon_learnset_url)

pokemon_infocards = soup.findAll('div',{'class':'infocard'})

pokemon_learnset_urls =[]

for p in pokemon_infocards:
    pokemon_learnset_urls.append('https://pokemondb.net/'+p.find('a').get('href')+'/moves/2')

pokemon_name =[]
move_name = []
move_power =[]
for pokemon in pokemon_learnset_urls:
#pokemon = pokemon_learnset_urls[0]
    time.sleep(0.5)
    r=requests.get(pokemon)
    pokemon_moves_df_list =[]

    for i in pd.read_html(r.text):
        tables = i['Move'].values
        for t in tables:
            move_name.append(t.lower())
            pokemon_name.append(pokemon.split('/')[5])
        power = i['Power'].values
        for p in power:
            move_power.append(p)
    
pokemon_learnset_df = pd.DataFrame()
pokemon_learnset_df['pokemon']=pokemon_name
pokemon_learnset_df['move_name']=move_name
pokemon_learnset_df['move_power']=move_power


pokemon_learnset_df.drop_duplicates(inplace=True)

pokemon_learnset_df['move_power']=pokemon_learnset_df.move_power.str.replace('—','0').astype(float)

pokemon_learnset_df.to_pickle(pokemon_learnset_df_picklename)

pokemon_learnset_df.head(5)
```

## Find non matching moves between datasets

```python
pokemon_moves_df = pokemon_moves_df.loc[pokemon_moves_df['move_name']!='struggle']

not_in_learnset =[]
not_in_moves =[]

print('Values not in pokemon_learnset_df:')
for i in pokemon_moves_df['move_name'].unique():
    if i not in pokemon_learnset_df['move_name'].unique():
        not_in_learnset.append(i)
        print(i)

print('\nValues not in pokemon_moves_df:')
for i in pokemon_learnset_df['move_name'].unique():
    if i not in pokemon_moves_df['move_name'].unique():
        not_in_moves.append(i)
        print(i)


print('\n Mappings list')
mappings = {}
for m,v in zip(not_in_learnset,not_in_moves):
    mappings[m]=v

print(mappings)
pokemon_moves_df.replace(mappings,inplace=True)
pokemon_moves_df.to_pickle(pokemon_moves_df_picklename)

pokemon_moves_df= pd.read_pickle(pokemon_moves_df_picklename)

print('Values not in pokemon_learnset_df:')
for i in pokemon_moves_df['move_name'].unique():
    if i not in pokemon_learnset_df['move_name'].unique():
        not_in_learnset.append(i)
        print(i)

print('\nValues not in pokemon_moves_df:')
for i in pokemon_learnset_df['move_name'].unique():
    if i not in pokemon_moves_df['move_name'].unique():
        not_in_moves.append(i)
        print(i)
    
```

```python
pokemon_stats_df=pokemon_stats_df.sort_values(by='pokemon')
pokemon_learnset_df =  pokemon_learnset_df.sort_values(by='pokemon')
not_in_learnset =[]
not_in_stats =[]

print('Values not in pokemon_learnset_df:')
for i in pokemon_stats_df['pokemon'].unique():
    if i not in pokemon_learnset_df['pokemon'].unique():
        not_in_learnset.append(i)
        print(i)

print('\nValues not in pokemon_moves_df:')
for i in pokemon_learnset_df['pokemon'].unique():
    if i not in pokemon_stats_df['pokemon'].unique():
        not_in_stats.append(i)
        print(i)
        
print('\n Mappings list')
mappings = {}
for m,v in zip(not_in_learnset,not_in_stats):
    mappings[m]=v

print(mappings)
pokemon_stats_df.replace(mappings,inplace=True)
pokemon_stats_df.to_pickle(pokemon_stats_df_picklename)

pokemon_stats_df= pd.read_pickle(pokemon_stats_df_picklename)

print('Values not in pokemon_learnset_df:')
for i in pokemon_stats_df['pokemon'].unique():
    if i not in pokemon_learnset_df['pokemon'].unique():
        not_in_learnset.append(i)
        print(i)

print('\nValues not in pokemon_moves_df:')
for i in pokemon_learnset_df['pokemon'].unique():
    if i not in pokemon_stats_df['pokemon'].unique():
        not_in_stats.append(i)
        print(i)
```

```python
damage = []
move_types =[]
move_categories=[]
for pokemon,move,move_power in zip(pokemon_learnset_df['pokemon'],pokemon_learnset_df['move_name'],pokemon_learnset_df['move_power']):
    pokemon_data = pokemon_stats_df.loc[pokemon_stats_df['pokemon']==pokemon]
    pokemon_attack = pokemon_data['attack'].values[0]
    pokemon_special_attack =pokemon_data['special_attack'].values[0]
    pokemon_types =[pokemon_data['main_type'].values[0],pokemon_data['secondary_type'].values[0]]

    move_type = pokemon_moves_df.loc[pokemon_moves_df['move_name']==move]['move_type'].values[0]
    #move_power = pokemon_moves_df.loc[pokemon_moves_df['move_name']==move]['move_power'].values[0]
    move_category = pokemon_moves_df.loc[pokemon_moves_df['move_name']==move]['move_category'].values[0]

    if move_category =='Physical':
        if move_type in pokemon_types:
            damage.append(((move_power*1.5)*pokemon_attack))
        else:
            damage.append((move_power*pokemon_attack))
    if move_category =='Special':
        if move_type in pokemon_types:
            damage.append(((move_power*1.5)*pokemon_special_attack))
        else:
            damage.append((move_power*pokemon_attack))
    if move_category =='Status':
        damage.append(0)
    move_types.append(move_type)
    move_categories.append(move_category)
    
pokemon_learnset_df['damage']=damage
pokemon_learnset_df['move_type']=move_types
pokemon_learnset_df['move_category']=move_categories

pokemon_learnset_df.to_pickle(pokemon_learnset_df_picklename)
```

# Interactive Visualizations

## Pokemon Stats

```python
n = 20
pokemon_type = ['all']
pokemon_type.extend(list(pokemon_stats_df['main_type'].sort_values().unique()))

def pokemon_stats_by_type(n,pokemon_type):
    plt.style.use('bmh')
    fig,ax=plt.subplots(2,1,figsize=(16,12))

    pokemon_stats_df= pd.read_pickle(pokemon_stats_df_picklename)

    if pokemon_type =='all':
        top_pokemon_total = pokemon_stats_df.nlargest(columns='total',n=n)[['pokemon','total']]
        ax[0].barh(top_pokemon_total['pokemon'],top_pokemon_total['total'])
        ax[0].invert_yaxis()

        ax[1].hist(pokemon_stats_df['total'],bins=20)
        ax[1].set_xticks([pokemon_stats_df['total'].min()+(25*(i)) for i in range(20)])

    else:
        pokemon_df = pokemon_stats_df.loc[(pokemon_stats_df['main_type']==pokemon_type)|(pokemon_stats_df['secondary_type']==pokemon_type)]
        top_pokemon_total =pokemon_df.nlargest(columns='total',n=n)[['pokemon','total']]
        ax[0].barh(top_pokemon_total['pokemon'],top_pokemon_total['total'])
        ax[0].invert_yaxis()

        ax[1].hist(pokemon_df['total'],bins=20)
        ax[1].set_xticks([pokemon_df['total'].min()+(25*(i)) for i in range(20)])


    plt.show()
    
w = interactive(pokemon_stats_by_type,n=[5,10,15,20,25],pokemon_type=pokemon_type)
display(w)
```
